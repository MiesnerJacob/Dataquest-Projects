# Data Cleaning

```python
star_wars = star_wars.iloc[1:]
```

```python
star_wars = star_wars[pd.notnull(star_wars["RespondentID"])]
```

Removing all rows that dont have an ID

```python
yes_no = {
    "Yes": True,
    "No": False,
    " ": np.nan
}
```

```python
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)
```

```python
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)
```

```python
print(star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].unique())
print(star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].unique())
```

```python
movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True,
        np.nan: False
}
```

# Data Analysis

## Top Rated Movie

```python
star_wars[star_wars.columns[9:15]].mean()
```

```python
plt.bar(range(6), star_wars[star_wars.columns[9:15]].mean())
```

After looking at the average rankings for each of the six moovies (lowest value equals best), we can see that Episode IV: The Empire Strikes Back was rated the best among fans who took the survey.

## Most Watched Movie

```python
star_wars[star_wars.columns[3:9]].sum()
```

```python
plt.bar(range(6), star_wars[star_wars.columns[3:9]].sum())
```

We can see from the chart above that Episode IV: The Empire Strikes Back was also the most viewed moovie, which may contribute to why it is the highest ranked amongst survey takers. Out of curiosity lets take a look at how many respondents consider themselves fans.

```python
star_wars[star_wars.columns[1]].value_counts()
```

It looks like most of the survey takers consider themselves to be Star Wars Fans, perhaps the reason some of the movies had been viewed more than others are due to the non-fans who have only seen some of the films.
